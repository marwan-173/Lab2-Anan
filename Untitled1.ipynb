```python
# @title Students Viewer + Favorite Program Updater (auto names from file)

import json
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---- File path ----
FILE_PATH = Path("/content/drive/My Drive/Colab Notebooks/students.json")

# ---- Mount Drive if needed ----
if not FILE_PATH.exists():
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except Exception as e:
        print("Couldn't auto-mount Drive. Run manually if needed.")

if not FILE_PATH.exists():
    raise FileNotFoundError("⚠️ students.json not found at /content/drive/My Drive/Colab Notebooks/")

# ---- Load JSON ----
with open(FILE_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

# ---- Build dropdown from names ----
names = [s["first_name"] for s in data]
dropdown = widgets.Dropdown(
    options=names,
    description='Student:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='50%')
)

# ---- Textbox for favorite program ----
fav_box = widgets.Text(
    placeholder='Enter new favorite program...',
    description='Program:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='50%')
)

# ---- Button ----
update_button = widgets.Button(
    description="Update",
    button_style='success',
    layout=widgets.Layout(width='20%')
)

# ---- Output area ----
output = widgets.Output()

def show_student_info(name):
    """Show student info in output."""
    with output:
        clear_output()
        student = next((s for s in data if s["first_name"].lower() == name.lower()), None)
        if not student:
            print("❌ Student not found.")
            return

        print(f"Name: {student['first_name']} {student['last_name']}")
        print("Email:", student["email"])
        print("Courses:")
        for c in student["courses"]:
            print(" •", c)
        print("Interesting link:", student["interesting_link"])
        print("Favorite program:", student.get("favorite_program", "(none)"))

def update_student(_):
    """Update favorite program and save to JSON."""
    name = dropdown.value
    fav = fav_box.value.strip()
    with output:
        clear_output()
        student = next((s for s in data if s["first_name"].lower() == name.lower()), None)
        if not student:
            print("❌ Student not found.")
            return

        if fav:
            student["favorite_program"] = fav
            with open(FILE_PATH, "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=2)
            print(f"✅ Updated favorite program for {student['first_name']} {student['last_name']}: {fav}")
        else:
            print(f"ℹ️ No new program entered — showing info for {student['first_name']} {student['last_name']}:")
        # Show student info after action
        print("\n--- Student Info ---")
        print(f"Name: {student['first_name']} {student['last_name']}")
        print("Email:", student["email"])
        print("Courses:")
        for c in student["courses"]:
            print(" •", c)
        print("Interesting link:", student["interesting_link"])
        print("Favorite program:", student.get("favorite_program", "(none)"))

# ---- Connect events ----
dropdown.observe(lambda change: show_student_info(change["new"]), names='value')
update_button.on_click(update_student)

# ---- Display UI ----
display(dropdown, fav_box, update_button, output)

# ---- Show first student initially ----
show_student_info(names[0])
```
